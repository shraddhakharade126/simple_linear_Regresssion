```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# import dataset

ds=pd.read_csv("dataset1.csv")
ds
```

```python
ds.columns
```

```python
# Divide dataset into independent and dependent variables

x=ds[['Study Hours']]
y=ds[['Exam Score']]
```

```python
# Plot

plt.scatter(x,y)
plt.show()
```

```python
# spliting dataset as training and Testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
# Introduce SLR MOdel

from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
# Predict

result= reg.predict([[10]])
result
```

```python
y_pre =reg.predict(x_test)
y_pre
```

```python
# checking accuracy
from sklearn.metrics import r2_score
score= r2_score(y_test,y_pre)
score
```

```python
# visulation

plt.scatter(x,y)
result =reg.predict(x)
plt.plot(x,result,color='red')
plt.show()
```

```python
# For User Input

study_hour =int(input("Enter Study hours : "))
myinput =pd.DataFrame(columns =['Study Hours'],data=[[study_hour]])
result =reg.predict(myinput)
print("Your score must be close to ", result[0,0])
```
